```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
brackets = ['https://animebracket.com/results/best-girl-contest-2014?group=full',
           'https://animebracket.com/results/best-girl-part-3-waifu-crusaders?group=full',
           'https://animebracket.com/results/best-girl-contest-2-22-the-salt-will-not-stop-coming-?group=full',
           'https://animebracket.com/results/best-girl-4-a-certain-salty-railgun?group=full']
```

```python
header = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/53.0.2785.143 Safari/537.36'}
```

```python
rows = {
    'ids1': [],
    'ids2': [],
    'votes1': [],
    'votes2': [],
    'images1': [],
    'images2': [],
    'tiers': [],
    'result': []
}
```

```python
total = 0
for url in brackets:
    print(url)
    r = requests.get(url,header)
    soup = BeautifulSoup(r.text,'lxml')
    bracket = json.loads(soup.script.text.split("window.bracketData = ")[1].rstrip()[:-1])['results']
    
    for tier in bracket:
        total += len(tier)
        print(len(tier))
        for matchup in tier:
            c1 = matchup['character1']
            c2 = matchup['character2']
            if matchup['final']:
                rows['ids1'].append(c1['id'])
                rows['ids2'].append(c2['id'])
                rows['votes1'].append(c1['votes'])
                rows['votes2'].append(c2['votes'])
                rows['images1'].append(c1['image'])
                rows['images2'].append(c2['image'])
                rows['tiers'].append(matchup['tier'])
                rows['result'].append(1 if c1['votes']>=c2['votes'] else 0)
            else:
                rows['ids1'].append(c1['id'])
                rows['ids2'].append(c2['id'])
                rows['votes1'].append(int(matchup['character1Votes']))
                rows['votes2'].append(int(matchup['character2Votes']))
                rows['images1'].append(c1['image'])
                rows['images2'].append(c2['image'])
                rows['tiers'].append(matchup['tier'])
                rows['result'].append(1 if rows['votes1'][-1]>=rows['votes2'][-1] else 0)
    
```

```python
for key in rows:
    print(key, len(rows[key]))
```
